```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
%pylab inline
plt.style.use('ggplot')

import glob
```

```python
colors = {0: "black", 20: "blue", 40: "green", 60: "yellow", 80: "orange", 100: "red"}

for g in glob.glob("../Datafiles/working/feature/shift_*.csv"):
    df = pd.read_csv(g)
    
    labels = df["confidence"].values
    std = df["std_temp"].values
    slope = df["slope_temp"].values
    
    plt.figure()
    plt.title(g)
    for i in range(len(labels)):
        plt.plot(std[i], slope[i], "o", color=colors[labels[i]])
        plt.xlim(0.0, 0.25)
```
